<h2>Sublime Limes' Line Graphs</h2>
<p>People all over the world can use this product to get the freshest, best-quality limes delivered to their door. One of the product managers at Sublime Limes would like to gain insight into the customers and their sales habits. Using Matplotlib, youâ€™ll create some different line graphs (or maybe, lime graphs) to communicate this information effectively.</p>

```python
from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Total visits
visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# Numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]


# Figure Size
plt.figure(figsize=(12, 8))

#Subplot Ax1
ax1 = plt.subplot(1,2,1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker = 'o')
plt.xlabel('Months')
plt.ylabel('Vists')
plt.title('Visits Per Month')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

#Subplot Ax2
ax2 = plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month, color = 'Red')
plt.plot(x_values, persian_limes_per_month,  color = 'Blue')
plt.plot(x_values, blood_limes_per_month,  color = 'Purple')
plt.legend(['Key Limes', 'Persian Limes', 'Blodd Limes'])
plt.title('Lime Sales Per Month')
plt.xlabel('Months')
plt.ylabel('Sales')
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.subplots_adjust(wspace = 0.3)

plt.show()
plt.clf()

#plt.savefig('Lime Species Grahps.png')
```
