```python
import pandas as pd

visits = pd.read_csv('visits.csv',
                parse_dates=[1])
cart = pd.read_csv('cart.csv',
                parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                parse_dates=[1])

head_of_visits = visits.head(11)
head_of_cart = cart.head(11)
head_of_checkouts = checkout.head(11)
head_of_purchases = purchase.head(11)
#print(head_of_visits,head_of_cart,head_of_checkouts,head_of_purchases)

# Merging of Vists/Cart
visits_and_cart = pd.merge(visits, cart, how ='left')
#visits.merge(cart, how='left')
print(visits_and_cart)
length_of_visitsandcart = len(visits_and_cart)
null_count_of_visitsandcart = visits_and_cart[visits_and_cart.cart_time.isnull()]
print(len(null_count_of_visitsandcart))
percentage_of_visitsandcart = float(len(null_count_of_visitsandcart)/length_of_visitsandcart)*100
print(percentage_of_visitsandcart)

# Merging of Cart/Checkout
cart_and_checkout = pd.merge(cart, checkout, how ='left')
#cart.merge(checkout, how='left')
print(cart_and_checkout)
length_of_cartcheckout = len(cart_and_checkout)
null_count_of_cartcheckout = cart_and_checkout[cart_and_checkout.checkout_time.isnull()]
print(len(null_count_of_cartcheckout))
percentage_of_cartcheckout = float(len(null_count_of_cartcheckout)/length_of_cartcheckout)*100
print(percentage_of_cartcheckout)

# Merging all
all_data = visits.merge(cart, how='left')\
                .merge(checkout, how='left')\
                .merge(purchase, how='left')
print(all_data)
print(all_data[all_data.checkout_time.isnull()])
alldata_percentile = float(len(all_data[all_data.checkout_time.isnull()])/len(all_data))*100
print(alldata_percentile)
weakest = all_data.max()
all_data['time_to_purchase'] = all_data.purchase_time - all_data.visit_time
print(all_data.time_to_purchase)
overall_average = all_data.time_to_purchase.mean()
print(overall_average)








```
