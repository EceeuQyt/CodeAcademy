```python
import pandas as pd

inventory = pd.read_csv('inventory.csv')
last_row = inventory.tail(1)
deca_rows = inventory.head(10)
print(last_row, deca_rows)
print()

staten_island = inventory.iloc[:10]
product_request = staten_island.product_description
print(staten_island)
print(product_request)
print()

brooklyn_seed_request = inventory[(inventory.location == 'Brooklyn') & (inventory.product_type == 'seeds')]  
print(brooklyn_seed_request)
print()

inventory['In Stock?'] = lambda x: True if inventory.quantity > 0 else False 
print(inventory.columns)
print()

inventory['Total_Value'] = inventory.price * inventory.quantity
print(inventory.Total_Value)
print()

combine_lambda = lambda row: f'{row.product_type} - {row.product_description}'
inventory['Full_Description'] = inventory.apply(combine_lambda, axis = 1)
print(inventory)
```
